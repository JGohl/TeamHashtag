```python
# dependencies
import csv
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from datetime import datetime
from collections import Counter
```

```python
# read csv file
# 5 10 12 receive an error message
file_data = ("IRAhandle_tweets_13.csv")
data = pd.read_csv(file_data)
data.head()
```

```python
# build data frame with selected columns
troll_df = pd.DataFrame(data[['external_author_id', 
                              'author', 
                              'content', 
                              'region', 
                              'language', 
                              'publish_date',
                              'following', 
                              'followers',
                             'account_type',
                             'account_category',
                             'retweet',
                             'tweet_id']])


# make publish date field a date field
troll_df['publish_date'] = troll_df['publish_date'].str.split(' ', expand=True)
troll_df['publish_date'] = pd.to_datetime(troll_df['publish_date'])
troll_df.head()
```

```python
# add date period column
#troll_df['time_period'] = ''
#troll_df.head()

# add data to period column based on publish date
  
# Attempt
#    if troll_df.loc[index, 'publish_date'] < pd.to_datetime('7/1/2015'):
#       troll_df.loc[index, 'time_period'] = 'Before analysis'
#    elif troll_df.loc[index, 'publish_date'] < pd.to_datetime('3/1/2016'):
#        troll_df.loc[index, 'time_period'] = 'Pre-election'
#    elif troll_df.loc[index, 'publish_date'] < pd.to_datetime('12/1/2016'):
#        troll_df.loc[index, 'time_period'] = 'Election season'
#    elif troll_df.loc[index, 'publish_date'] < pd.to_datetime('7/1/2017'):
#        troll_df.loc[index, 'time_period'] = 'Post-election'
#    else:
#        troll_df.loc[index, 'time_period'] = 'After analysis'

before = (troll_df['publish_date'] > pd.to_datetime('6/30/2015')) & (troll_df['publish_date'] <= pd.to_datetime('3/1/2016')) 
during = (troll_df['publish_date'] > pd.to_datetime('3/1/2016'))  & (troll_df['publish_date'] <= pd.to_datetime('12/1/2016'))
after = (troll_df['publish_date'] > pd.to_datetime('12/1/2016')) & (troll_df['publish_date'] < pd.to_datetime('7/1/2017'))

troll_df.loc[before, "time_period"] = 'Pre-election'
troll_df.loc[during, "time_period"] = 'Election season'
troll_df.loc[after, "time_period"] = 'Post-election'

troll_df.groupby('time_period').count()
```

```python
# keep only rows where the language is English
english_troll = troll_df[troll_df['language'].str.match('English')]
english_troll.index = pd.RangeIndex(len(english_troll.index))
english_troll.head()
```

```python
# find most used words in content


words = []
for row in english_troll:
    csv_words = row[2].split(" ")
    for i in csv_words:
        words.append(i)

words_counted = []
for i in words:
    x = words.count(i)
    words_counted.append((i,x))

set(words_counted)
    
print(words_counted)




#content = (row[2] for row in english_troll)
#print("Most frequent value: {0}".format(Counter(content).most_common()[0][0]))
```
