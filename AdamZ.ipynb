```python
#Libraries and Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import datetime
```

```python
#point to the csv
file_to_load = 'IRAhandle_tweets_13.csv'
#read csv into a pandas dataframe
df_file = pd.read_csv(file_to_load)
df_file.head(2)
```

```python
df_data = pd.DataFrame(df_file[['tweet_id',
                   'external_author_id',
                   'author',
#                  'content',
#currently don't care about content
                   'region',
                   'language',
                   'publish_date',
                   'following',
                   'followers',
                   'account_type',
                   'retweet',
                   'account_category'
                  ]])
df_data.head()
```

```python
df_data['publish_date'] = df_data['publish_date'].str.split(' ', expand=True)
df_data.head()
```

```python
#I want to create a Column for election_timeframe
#Options being before, durring, and after
#The cutoff for before will be the February 1st 2016, the first day of primaries for 2016 election
#The cutoff for after will be after inauguration. January 20th 2017
df_data.loc[:,'publish_date'] = pd.to_datetime(df_data.loc[:,'publish_date'])
#gives an error, but stil evaluated. the publish_date column is now a pandas timestamp
before = (df_data['publish_date'] > pd.to_datetime('6/30/2015')) & (df_data['publish_date'] <= pd.to_datetime('3/1/2016')) 
during = (df_data['publish_date'] > pd.to_datetime('3/1/2016'))  & (df_data['publish_date'] <= pd.to_datetime('12/1/2016'))
after = (df_data['publish_date'] > pd.to_datetime('12/1/2016')) & (df_data['publish_date'] < pd.to_datetime('7/1/2017'))

df_data.loc[before, "time_period"] = 'Pre-election'
df_data.loc[during, "time_period"] = 'Election season'
df_data.loc[after, "time_period"] = 'Post-election'
df_data.head()
```
